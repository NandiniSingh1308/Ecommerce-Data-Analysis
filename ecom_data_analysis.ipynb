```python
ndf=df.dropna()
```

```python
ndf.head()
```

```python
ndf.info()
```

```python
ndf.isnull().sum()
```

```python
ndf["CustomerID"]=ndf["CustomerID"].astype("int64") #changing the datatype of the column from flaot to int
```

```python
ndf.head()
```

```python
ndf.describe()
```

```python
ndf.describe().round(2)
```

```python
ndf=ndf[ndf.Quantity>0] #removing negative values in quantity column
```

```python
ndf.info()
```

```python
ndf["Total_Cost"]=ndf["Quantity"]*ndf["UnitPrice"]
```

```python
ndf.head()
```

```python
ndf=ndf[["InvoiceNo","StockCode","Description","InvoiceDate","CustomerID","Country","Quantity","UnitPrice","Total_Cost"]] #rearranging the columns
```

```python
ndf.head()
```

```python
#to find monthly sales record-
```

```python
ndf['InvoiceDate'] = pd.to_datetime(ndf.InvoiceDate, format='%m/%d/%Y %H:%M') #the datatype of InvoiceDate was strings. Using pandas function "to_datetime", we are converting the string datatype to datetime datatype. We made it datetime because the pandas can determine the month and date and we could store it. Other way of doing it would have been string slicing since the datatype was string. But pandas is the elegant one.
```

```python
ndf.insert(loc=4,column="Month",value=ndf.InvoiceDate.dt.month) # we are extrcating data from ndf from the column InvoiceDate and it contains a datetime object and we need month hence, dt.month, if we needed year, dt.year etc.
```

```python
ndf.head()
```

```python
ndf.head()
```

```python
ndf.insert(loc=5,column="Day",value=ndf.InvoiceDate.dt.dayofweek+1) 
```

```python
ndf.head()
```

```python
ndf.insert(loc=5,column="Hour",value=ndf.InvoiceDate.dt.hour)
```

```python
ndf.head()
```

# Exploratory Data Analysis


```python
# how many orders are made by the customers
orders= ndf.groupby(by=["CustomerID","Country"],as_index=False)["InvoiceNo"].count().head(n=10)
```

```python
orders
```

```python
# top 10 customers
print("The top 10 customers with the most number of orders")
orders.sort_values(by="InvoiceNo",ascending=False).head(n=10)
```

```python
plt.subplots(figsize=(20,10))
plt.plot(orders.CustomerID,orders.InvoiceNo)
plt.xlabel("CustomerID")
plt.ylabel("Number of orders")
plt.title("Total number of orders for different customers")
plt.show()
```

```python
money_spent=ndf.groupby(by=["CustomerID","Country"],as_index=False)["Total_Cost"].sum().head(n=10)
```

```python
print("The top 10 highest spending customers")
money_spent.sort_values(by="Total_Cost",ascending=False).head(n=10)
```

```python
plt.subplots(figsize=(20,10))
plt.plot(money_spent.CustomerID,money_spent.Total_Cost)
plt.xlabel("CustomerID")
plt.ylabel("Total Cost")
plt.title("Top 10 highest spending Customers")
plt.show()
```

```python
ndf.insert(loc=5,column="year_month",value=ndf["InvoiceDate"].map(lambda x: 100*x.year + x.month))
```

```python
ndf.head()
```

```python
# Save a palette to a variable:
palette = sns.color_palette("bright")
# Use palplot and pass in the variable:
sns.palplot(palette)
```

```python
#ax=ndf.groupby("InvoiceNo") with a new function called unique()
#ax=ndf.groupby("Total_Cost")["Country"].unique().value_counts().sort_index().plot("bar",color=palette.any(),figsize=(15,6))
ax= ndf.groupby(by=["Country"],as_index=False)["InvoiceNo"].count().head(n=10)
```

```python
ax.sort_values(by="InvoiceNo",ascending=False).head(n=10)
```

```python
sns.set(rc={'figure.figsize':(20,15)})
a=sns.barplot(x="Country",y="InvoiceNo",data=ax)
a.set(xlabel="Country", ylabel = "Total Number of Orders")
plt.show()
```

```python
money_spent2=ndf.groupby(by=["Country"],as_index=False)["Total_Cost"].sum().head(n=10)
money_spent2.sort_values(by="Total_Cost",ascending=False).head(n=10)
sns.set(rc={'figure.figsize':(20,15)})
b=sns.barplot(x="Country",y="Total_Cost",data=money_spent2)
b.set(xlabel="Country", ylabel = "Total sales")
plt.show()
```

```python
orders2= ndf.groupby(by=["year_month"],as_index=False)["InvoiceNo"].count().head(n=13)
orders2.sort_values(by="year_month").head(n=13)
```

```python
sns.set(rc={'figure.figsize':(20,15)})
c=sns.barplot(x="year_month",y="InvoiceNo",data=orders2)
c.set(xlabel="Month", ylabel = "Total sales")
c.set_xticklabels(('Dec_10','Jan_11','Feb_11','Mar_11','Apr_11','May_11','Jun_11','July_11','Aug_11','Sep_11','Oct_11','Nov_11','Dec_11'), rotation='horizontal', fontsize=13)
plt.show()
```

```python
orders3= ndf.groupby(by=["Day"],as_index=False)["InvoiceNo"].count()
orders3.sort_values(by="Day")
sns.set(rc={'figure.figsize':(20,15)})
d=sns.barplot(x="Day",y="InvoiceNo",data=orders3)
d.set(xlabel="Day", ylabel = "Total sales")
d.set_xticklabels(('Monday','Tuesday','Wednesday','Thursday','Friday','Sunday'), rotation='horizontal', fontsize=13)
plt.show()
```
